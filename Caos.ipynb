## Dinâmica de populações

### A equação que captura a essência deste processo é:
$ y_{n+1}=ry_{n}(1-y_n)$


* Faça gráficos de $y_n$ como função de $n$ para $r=0,5$. Use $y_0=0,5$ para $n$ entre 1 e 50.

```python
import matplotlib.pyplot as plt

def log_fun(r, len_n, y0):
    yn = [y0]
    for i in range(1, len_n+1):
        yn.append(r*yn[i-1]*(1-yn[i-1]))
    return yn

y1 =log_fun(0.5, 50, 0.5)
y2 =log_fun(2.5, 50, 0.5)
y3 =log_fun(3.1, 50, 0.5)
y4 =log_fun(3.5, 50, 0.5)
y5 =log_fun(3.7, 50, 0.5)

n = [i for i in range(51)]

def singplot(y1,y2,y3,y4,y5, n, valor_inicial):
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(n, y1, label='r=0.5') 
    ax.plot(n, y2, label='r=2.5') 
    ax.plot(n, y3, label='r=3.1') 
    ax.plot(n, y4, label='r=3.5')  
    ax.plot(n, y5, label='r=3.7')
    ax.set_xlabel('n', fontsize="15") 
    ax.set_ylabel('y', fontsize="15")  
    ax.set_title(f"Diferentes valores de $r$\n Para $y_0={valor_inicial}$", fontsize="22")  
    ax.legend()  # Add a legend. 
    
singplot(y1,y2,y3,y4,y5, n, "0.5")
```

* Refazer considerando as seguintes três condições iniciais:
    * $y_0=0.5$, $y_0=0.501$, $y_0=0.5001$

```python
y0s = [0.5, 0.501, 0.5001]
for i in range(len(y0s)):
    x1 =log_fun(0.5, 50, y0s[i])
    x2 =log_fun(2.5, 50, y0s[i])
    x3 =log_fun(3.1, 50, y0s[i])
    x4 =log_fun(3.5, 50, y0s[i])
    x5 =log_fun(3.7, 50, y0s[i])
    singplot(x1,x2,x3,x4,x5, n, y0s[i])
    print()
```

* Considere 1000 valores de $r$ igualmente espaçados entre $10^{-5}$ e $4$. Para cada um desses valores de $r$, iteramos a equação logística 1000 vezes ($n_{max}=1000$). Em um gráfico de $y_n$ como função de $r$, vamos plotar para cada valor de $r$ os 100 últimos passos da evolução. 

```python
import numpy as np


def diagrama(r, n=[i for i in range(1000)], ymin=-0.02, ymax=1):
    y = []
    
    for item in r:
        yn = log_fun(item, len(n), 0.5)    
        yn_t = tuple(yn[901:])
        y.append(yn_t)

    plt.figure(figsize=(12, 6), dpi=80)
    plt.title("Diagrama de bifurcação", fontsize="22")
    plt.ylim([ymin, ymax])
    plt.plot(r, y)
    plt.show()
    return

r = np.linspace(10**(-5), 4.0, 1000)
n = [i for i in range(1000)]

diagrama(r, n)
```
